```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import StaleElementReferenceException
from tqdm import tqdm
import time
import pandas as pd
import os

driver = webdriver.Chrome()
driver.get('https://www.google.com/maps')
time.sleep(3)


def fetch_place_details(driver, category):
    places = []

    location_elements = driver.find_elements(By.CLASS_NAME, 'Nv2PK')

    for element in location_elements:
        try:
    
            rating = element.find_element(By.CLASS_NAME,'MW4etd').text if len(element.find_elements(By.CLASS_NAME,'MW4etd')) > 0 else 'N/A'
            reviews = element.find_element(By.CLASS_NAME,'UY7F9').text if len(element.find_elements(By.CLASS_NAME,'UY7F9')) > 0 else 'N/A'
            contact = element.find_element(By.CLASS_NAME,'UsdlK').text if len(element.find_elements(By.CLASS_NAME,'UsdlK')) > 0 else 'N/A'
            element.click()
            time.sleep(3)              
            name = driver.find_element(By.CLASS_NAME, 'DUwDvf').text if len(driver.find_elements(By.CLASS_NAME, 'DUwDvf')) > 0 else 'N/A'
            address = driver.find_element(By.CLASS_NAME, 'Io6YTe').text if len(driver.find_elements(By.CLASS_NAME, 'Io6YTe')) > 0 else 'N/A'
            url=driver.current_url
            places.append({
                'name': name,
                'address': address,
                'type_of_place': category,
                'rating': rating,
                'reviews': reviews,
                'contact': contact,
                'url':url
            })

            
            
            
        except StaleElementReferenceException as e:
            print(f"Stale element encountered")
            continue
        except Exception as e:
            print(f"NO INFORMATION FOR THIS LOCATION AVAILABLE")
            time.sleep(1)
            continue

    return places


```

```python
import pyautogui
def scroll():
    pyautogui.scroll(-5000)
    time.sleep(3)
    pyautogui.scroll(-5000)
    time.sleep(3)
    pyautogui.scroll(-5000)
    time.sleep(3)
    pyautogui.scroll(-5000)
    time.sleep(3)
    pyautogui.scroll(-5000)
    time.sleep(3)
    pyautogui.scroll(-5000)
    time.sleep(3)
    pyautogui.scroll(-5000)
    time.sleep(3)
    pyautogui.scroll(-5000)
    time.sleep(3)
    pyautogui.scroll(-8000)
    time.sleep(3)
    pyautogui.scroll(-8000)
    time.sleep(3)
```

```python

queries = ["warehouses in india", "malls in india", "Hostpitals in india","parking-lots in india","schools in india",'Universities in india']
all_places = []

for query in queries:
    
    search_box = driver.find_element(By.ID, "searchboxinput")
    search_box.clear()
    search_box.send_keys(query)
    search_box.send_keys(Keys.ENTER)
    time.sleep(5)
    
    scroll()
        
    places = fetch_place_details(driver, query)  # Pass the category as the query
    all_places.extend(places)
    
    search_box.clear()
    time.sleep(1) 


```
